```python
import geopandas as gpd

from google.colab import files
```

```python
South_Asia= gpd.read_file("/content/South_Asia_polygon2.shp")
print(South_Asia)
```

```python
import matplotlib.pyplot as plt
from matplotlib_scalebar.scalebar import ScaleBar
```

```python
from matplotlib_scalebar.scalebar import ScaleBar

MapFrame = plt.subplots(figsize=(10, 10))
South_Asia.plot(column='Name',
                cmap='Blues_r',
                legend=False,
                edgecolor='black',
                linestyle='--',
                linewidth=1,
                figsize=(10, 10),
                ax=MapFrame[1] # Pass the axes object
                )

MapFrame[1].set_title('Bay of Bengal, Bangladesh', color='green',
                      fontsize=16, fontweight='bold')

x = 0.9
y = 0.9
arrow_length= 0.1
MapFrame[1].annotate('N',
            xy=(x, y), xytext=(x, y - arrow_length),
            arrowprops=dict(facecolor='black', width=5, headwidth=12),
            ha='center', va='center', fontsize=14, zorder=7,
            xycoords=MapFrame[1].transAxes)


scalebar = ScaleBar(1.0, location='center right')
MapFrame[1].add_artist(scalebar)

# Add gridlines
MapFrame[1].grid(False, which='both', color='gray', linestyle='--',
                 linewidth=0.5)

MapFrame[1].set_xlabel("Longitude", fontsize=12)
MapFrame[1].set_ylabel("Latitude", fontsize=12)
plt.show()
```

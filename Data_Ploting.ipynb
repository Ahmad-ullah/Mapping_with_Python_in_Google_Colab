```python
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
import pandas as pd
```

```python
from google.colab import files
District_Pop= pd.read_excel('Population_Data.xlsx')
District_Pop.head()
```

```python
#For a 1x2 plotting area
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))
f.subplots_adjust(wspace=0.2)

# For First Plot
ax1.bar(District_Pop.iloc[:, 0], District_Pop.iloc[:, 1])
ax1.set_title ('Population by District',
               color= 'b',
               fontsize=14,
               fontweight='bold',
               loc='center'
               )
ax1.set_xlabel('District')
ax1.set_ylabel('Population')
ax1.tick_params(axis='x', rotation=90)


# For Second Plot
ax2.set_title ('Second Plot',
               color= 'b',
               fontsize=14,
               fontweight='bold',
               loc='center'
               )
ax2.set_xlabel('X-axis Label')
ax2.set_ylabel('Y-axis Label')

plt.tight_layout()
plt.show()
```
